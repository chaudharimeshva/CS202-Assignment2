```python
import pandas as pd
import json, xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
import seaborn as sns

# CWE Top 25 list
CWE_TOP_25 = {'CWE-79','CWE-89','CWE-20','CWE-125','CWE-78','CWE-22','CWE-416','CWE-352',
              'CWE-434','CWE-476','CWE-287','CWE-190','CWE-502','CWE-269','CWE-863',
              'CWE-77','CWE-918','CWE-94','CWE-798','CWE-400','CWE-306','CWE-862',
              'CWE-276','CWE-427','CWE-732'}

# Function to normalize CWE IDs across tools
def normalize_cwe_id(cwe_id):
    if pd.isna(cwe_id):
        return None
    cwe_id = str(cwe_id).strip()
    if cwe_id.startswith("CWE-"):
        cwe_id = cwe_id.replace("CWE-", "")
    return cwe_id

# Load Bandit
def load_bandit(file, project):
    data = json.load(open(file))
    results = []
    for i in data.get("results", []):
        cwe_field = i.get("issue_cwe")
        if cwe_field and cwe_field.get("id"):
            cwe_ids = cwe_field["id"]
            # Handle list or string
            if isinstance(cwe_ids, list):
                for cwe in cwe_ids:
                    results.append({"Project": project, "Tool": "Bandit", "CWE_ID": normalize_cwe_id(cwe), "Findings": 1})
            else:
                results.append({"Project": project, "Tool": "Bandit", "CWE_ID": normalize_cwe_id(cwe_ids), "Findings": 1})
    return results

# Load Semgrep
def load_semgrep(file, project):
    data = json.load(open(file))
    results = []
    for match in data.get("results", []):
        cwe = match["extra"]["metadata"].get("cwe")
        if cwe:
            if isinstance(cwe, list):
                for c in cwe:
                    results.append({"Project": project, "Tool": "Semgrep", "CWE_ID": normalize_cwe_id(c), "Findings": 1})
            else:
                results.append({"Project": project, "Tool": "Semgrep", "CWE_ID": normalize_cwe_id(cwe), "Findings": 1})
    return results

# Load Cppcheck
def load_cppcheck(file, project):
    tree = ET.parse(file)
    root = tree.getroot()
    data = []
    for error in root.findall(".//error"):   # more robust XML path
        cwe = error.attrib.get("cwe")
        if cwe:
            data.append({"Project": project, "Tool": "Cppcheck", "CWE_ID": normalize_cwe_id(cwe), "Findings": 1})
    return data


# Define file paths (make sure filenames exist in same folder)
files = [
    # cJSON
    ("bandit_cJSON.json", "Bandit", "cJOSN"),
    ("semgrep_cJSON.json", "Semgrep", "cJOSN"),
    ("cppcheck_cJSON.xml", "Cppcheck", "cJOSN"),

    # copyparty
    ("bandit_copyparty.json", "Bandit", "copyparty"),
    ("semgrep_copyparty.json", "Semgrep", "copyparty"),
    ("cppcheck_copyparty.xml", "Cppcheck", "copyparty"),

    # openPilot
    ("bandit_openpilot.json", "Bandit", "openpilot"),
    ("semgrep_openpilot.json", "Semgrep", "openpilot"),
    ("cppcheck_openpilot.xml", "Cppcheck", "openpilot")
]

# Read all files
all_data = []

for f, tool, project in files:
    try:
        if tool == "Bandit":
            all_data += load_bandit(f, project)
        elif tool == "Semgrep":
            all_data += load_semgrep(f, project)
        elif tool == "Cppcheck":
            all_data += load_cppcheck(f, project)
        print(f" Loaded {tool} results for {project}")
    except Exception as e:
        print(f" Error loading {f}: {e}")

# Create DataFrame
df = pd.DataFrame(all_data)

# Normalize CWE_ID to numeric part only
df["CWE_ID"] = df["CWE_ID"].astype(str).str.extract(r'(\d+)')[0]  # extract just the number
df = df.dropna(subset=["CWE_ID"])  # drop rows where no CWE number found
df["CWE_ID"] = df["CWE_ID"].str.strip()

df["Is_In_CWE_Top_25"] = df["CWE_ID"].apply(lambda x: "Yes" if "CWE-" + str(x) in CWE_TOP_25 else "No")

df.to_csv("vulnerability_results_all.csv", index=False)

print("\n Combined data summary:")
print(df.head())
print("\nUnique Tools:", df["Tool"].unique())

# TOOL-LEVEL CWE COVERAGE
coverage = df.groupby("Tool")["CWE_ID"].nunique().reset_index()
coverage.columns = ["Tool", "Unique_CWEs"]

plt.figure(figsize=(7,5))
plt.bar(coverage["Tool"], coverage["Unique_CWEs"], color="skyblue")
plt.title("Unique CWE Coverage per Tool")
plt.xlabel("Tool")
plt.ylabel("Unique CWEs Found")
plt.show()

# TOP 25 CWE COVERAGE (%)
top25_df = df[df["Is_In_CWE_Top_25"] == "Yes"]
top25_coverage = top25_df.groupby("Tool")["CWE_ID"].nunique().reset_index()
top25_coverage.columns = ["Tool", "Top25_CWEs"]

summary = pd.merge(coverage, top25_coverage, on="Tool", how="left").fillna(0)
summary["Top25_Coverage_%"] = (summary["Top25_CWEs"] / len(CWE_TOP_25)) * 100
summary.to_csv("Tool_Level_CWE_Coverage.csv", index=False)

print("\n Tool-level CWE coverage summary:")
print(summary)

plt.figure(figsize=(7,5))
plt.bar(summary["Tool"], summary["Top25_Coverage_%"], color="lightcoral")
plt.title("CWE Top 25 Coverage (%) per Tool")
plt.xlabel("Tool")
plt.ylabel("Coverage Percentage (%)")
plt.show()

# PAIRWISE IOU ANALYSIS
tools = df["Tool"].unique()
iou_matrix = pd.DataFrame(index=tools, columns=tools)

for t1 in tools:
    set1 = set(df[df["Tool"] == t1]["CWE_ID"])
    for t2 in tools:
        set2 = set(df[df["Tool"] == t2]["CWE_ID"])
        iou = len(set1 & set2) / len(set1 | set2) if len(set1 | set2) > 0 else 0
        iou_matrix.loc[t1, t2] = iou

print("\n IoU Matrix:")
print(iou_matrix)

sns.heatmap(iou_matrix.astype(float), annot=True, cmap="Blues")
plt.title("Pairwise IoU between Tools (CWE IDs)")
plt.show()

# Save IoU matrix
iou_matrix.to_csv("IoU_Matrix.csv")
```
